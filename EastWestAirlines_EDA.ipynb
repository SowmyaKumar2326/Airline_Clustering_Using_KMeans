### Exploratory Data Analysis(EDA)

```python
sns.countplot(x=data.cc1_miles)
plt.title("Countplot of cc1_miles")
plt.savefig("Airline_Countplot1.png")
```

```python
data.cc1_miles.value_counts()
```

```python
sns.countplot(x=data.cc2_miles)
plt.title("Countplot of cc2_miles")
plt.savefig("Airline_Countplot2.png")
```

```python
data.cc2_miles.value_counts()
```

```python
sns.countplot(x=data.cc3_miles)
plt.title("Countplot of cc3_miles")
plt.savefig("Airline_Countplot3.png")
```

```python
data.cc3_miles.value_counts()
```

```python
sns.countplot(x=data.Bonus_trans,width = 0.7)
plt.title("Countplot of Bonus_trans")
plt.savefig("Airline_countplot4.png")
```

```python
sns.countplot(x=data.Flight_trans_12,width = 0.8)
plt.title("Countplot of Flight_trans_12")
plt.savefig("Airline_countplot5.png")
```

```python
sns.countplot(x="Award",width = 0.8,data=data)
plt.title("Free Flight")
plt.savefig("Countplot_freeflight.png")
```

```python
data.Award.value_counts()
```

```python
sns.heatmap(df[["Balance","Qual_miles","cc1_miles","cc2_miles","cc3_miles","Bonus_miles","Bonus_trans","Flight_miles_12mo","Flight_trans_12","Days_since_enroll","Award"]].corr(),annot=True)
plt.title("Heatmap")
plt.savefig("Airline_heatmap.png")
```

```python
sns.scatterplot(x=data.Qual_miles,y=data.Bonus_miles,hue=data.Award,data=data)
plt.title('Airline_Scatterplot')
plt.savefig("Airline_scatter1.png")
```

```python
sns.scatterplot(x=data.Days_since_enroll,y=data.Balance,hue=data.Award,data=data)
plt.title('Airline_Scatterplot')
plt.savefig("Airline_scatter2.png")
```
